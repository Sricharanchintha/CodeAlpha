```python
# 3: Where are most jobs located?
# Many jobs have multiple cities, so let's split and count all mentions
def extract_cities(location):
    if pd.isna(location):
        return []
    # Remove content in parentheses and split by comma
    cleaned = re.sub(r'\([^)]*\)', '', str(location))
    cities = [city.strip() for city in cleaned.split(',')]
    return [c for c in cities if c and 'Remote' not in c and 'Hybrid' not in c]

all_cities = df['Location'].apply(extract_cities).explode()
print("Top 10 Locations (counting all city mentions):")
print(all_cities.value_counts().head(10))
```

```python
# 4:Remote & Hybrid count
remote_count = df['Location'].astype(str).str.contains('Remote', case=False).sum()
hybrid_count = df['Location'].astype(str).str.contains('Hybrid', case=False).sum()
print(f"\nRemote jobs: {remote_count} ({remote_count/len(df)*100:.1f}%)")
print(f"Hybrid jobs: {hybrid_count} ({hybrid_count/len(df)*100:.1f}%)")
```

```python
# 5. Experience Analysis
def parse_exp_range(exp):
    if pd.isna(exp) or str(exp) in ['N/A']:
        return np.nan, np.nan
    nums = re.findall(r'\d+', str(exp))
    if len(nums) == 0:
        return np.nan, np.nan
    elif len(nums) == 1:
        return int(nums[0]), int(nums[0])
    else:
        return int(nums[0]), int(nums[1])

df[['min_exp', 'max_exp']] = df['Experience'].apply(
    lambda x: pd.Series(parse_exp_range(x))
)
df['avg_exp'] = (df['min_exp'] + df['max_exp']) / 2

print("\n=== Experience Statistics ===")
print(df[['min_exp', 'max_exp', 'avg_exp']].describe())

print("\nMost common experience ranges:")
print(df['Experience'].value_counts().head(10))
```

```python
# 6. Salary Analysis
print("\n=== Salary Disclosure ===")
total_jobs = len(df)
disclosed = df['Salary'].notna() & ~df['Salary'].astype(str).str.contains('Not Disclosed|Unpaid', case=False)
disclosed_count = disclosed.sum()
print(f"Jobs with disclosed salary: {disclosed_count} ({disclosed_count/total_jobs*100:.1f}%)")
```

```python
# Parse salary to numeric (annual in rupees)
def parse_salary(sal):
    if pd.isna(sal) or 'Not Disclosed' in str(sal) or 'Unpaid' in str(sal):
        return np.nan
    s = str(sal).replace(',', '').upper()
    if '/MONTH' in s:
        num = re.search(r'(\d+\.?\d*)', s)
        return float(num.group(1)) * 12 * 100000 if num else np.nan
    nums = re.findall(r'\d+\.?\d*', s)
    if not nums:
        return np.nan
    vals = [float(n) for n in nums]
    if max(vals) > 100:  # likely in rupees
        return np.mean(vals)
    else:  # in lacs
        return np.mean(vals) * 100000

df['salary_numeric'] = df['Salary'].apply(parse_salary)

print("\nSalary Statistics (in ₹ per year):")
print(df['salary_numeric'].describe())

print(f"Median salary: ₹{df['salary_numeric'].median():,.0f}")
print(f"Average salary: ₹{df['salary_numeric'].mean():,.0f}")
```
